````python
%sql
-- 🎯 Task:
-- For each month:

-- Find top 2 customers by total revenue
-- 🧠 Concepts You MUST Use:
-- GROUP BY
-- SUM(amount)
-- ROW_NUMBER()
-- PARTITION BY month
-- 🧠 Thinking Flow (VERY IMPORTANT)
-- Step 1 → Aggregate revenue per customer per month
-- Step 2 → Apply ROW_NUMBER() partition by month
-- Step 3 → Filter rank <= 2
-- ⚠️ Common Mistake (Avoid This)
-- Don’t try everything in one query.

-- Use CTE:

-- CTE 1 → aggregation
-- CTE 2 → ranking
-- Final → filter

with cte1 as(
  select 
 date_trunc('Month', order_date) as date_month ,customer_id , sum(amount) as revenue
from orders
group by  date_trunc('Month', order_date) ,customer_id
), cte2
(
  select * , dense_rank() over (partition by date_month order by revenue desc) as rank_no from cte1
)
select * from cte2
where rank_no <= 2



````
