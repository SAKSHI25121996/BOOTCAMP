## 🟢 Problem 2: Merge Two Dictionaries

### Input:


### Output:

```python
{"a":15, "b":20, "c":15}
```

---

````python
d1 = {"a": 10, "b": 20}
d2 = {"a": 5, "c": 15}

print("=== METHOD 1: Loop through KEYS only ===")
for key in d1:
    print(f"Key: {key}")
    print(f"Value: {d1[key]}")
    print()

print("=== METHOD 2: Loop through KEYS and VALUES together ===")
for key, value in d1.items():
    print(f"Key: {key}, Value: {value}")

print()
print("=== METHOD 3: Loop through just VALUES (rare) ===")
for value in d1.values():
    print(f"Value: {value}")
````

````python
d1 = {"a": 10, "b": 20}
d2 = {"a": 5, "c": 15}

result = {}

for key, value in d1.items():
    if key not in result:
        result[key] = value

for key , value in d2.items():
    if key not in result:
        result[key] = value
    else:
        result[key] += value

print(result)

                 
            
````

## 🟢 Problem 3: Flatten Nested List (Tricky)

### Input:
```python
data = [[1,2], [3,4], [5]]
```

### Output:
```python
[1,2,3,4,5]
```

### 🧠 THINK:
```
Outer list → loop through each sublist
Inner list → loop through each number
```

### Strategy:
1. Create an empty result list
2. Loop through each SUBLIST in data
3. For each sublist, loop through each NUMBER
4. Add each number to result

````python
data = [[1,2], [3,4], [5]]

result = []
for i in data:
    for x in i:
        if x not in result:
            result.append(x)
print(result)
````

## 🟢 SQL Problem: Increasing Revenue Customers

### Goal:
Find customers whose **revenue is increasing month-over-month**

### What we need:

Let's start! 👇

````python
%sql
-- Step 1: Drop table if it exists (so we can re-run this)
DROP TABLE IF EXISTS orders;

-- Step 2: Create orders table
CREATE TABLE orders (
    order_id INT,
    customer_id STRING,
    order_date DATE,
    amount DECIMAL(10,2)
);
````

````python
%sql
-- Insert sample data for different customers across 3 months
INSERT INTO orders VALUES
-- Customer A: INCREASING revenue (100 -> 150 -> 200)
(1, 'A', '2024-01-15', 100.00),
(2, 'A', '2024-02-10', 150.00),
(3, 'A', '2024-03-05', 200.00),

-- Customer B: DECREASING revenue (200 -> 150 -> 100)
(4, 'B', '2024-01-20', 200.00),
(5, 'B', '2024-02-15', 150.00),
(6, 'B', '2024-03-10', 100.00),

-- Customer C: MIXED (up then down)
(7, 'C', '2024-01-10', 100.00),
(8, 'C', '2024-02-20', 180.00),
(9, 'C', '2024-03-15', 120.00),

-- Customer D: INCREASING revenue (50 -> 75 -> 100)
(10, 'D', '2024-01-25', 50.00),
(11, 'D', '2024-02-18', 75.00),
(12, 'D', '2024-03-22', 100.00),

-- Customer E: Multiple orders in same month (should be aggregated)
(13, 'E', '2024-01-05', 50.00),
(14, 'E', '2024-01-28', 50.00),  -- Total Jan: 100
(15, 'E', '2024-02-12', 80.00),
(16, 'E', '2024-02-25', 70.00),  -- Total Feb: 150
(17, 'E', '2024-03-08', 120.00),
(18, 'E', '2024-03-20', 80.00);  -- Total Mar: 200
````

````python
%sql
-- Let's see all the orders
SELECT * FROM orders
ORDER BY customer_id, order_date;
````

````python
%sql
-- YOUR TASK: Write a query to find customers with INCREASING revenue
-- 
-- Expected customers: A, D, E
--
-- Steps:
-- 1. Calculate monthly revenue per customer (use SUM and DATE_TRUNC)
-- 2. Use LAG() to get previous month's revenue
-- 3. Filter where current_revenue > previous_revenue

-- Write your query here:
````

````python
query = """
WITH monthly_revenue AS (
    SELECT
        customer_id,
        DATE_TRUNC('month', order_date) AS month,
        SUM(amount) AS revenue
    FROM orders
    GROUP BY customer_id, month
),
revenue_diff AS (
    SELECT
        customer_id,
        month,
        revenue,
        LAG(revenue) OVER (PARTITION BY customer_id ORDER BY month) AS prev_revenue
    FROM monthly_revenue
),
increasing_customers AS (
    SELECT customer_id
    FROM revenue_diff
    WHERE prev_revenue IS NOT NULL AND revenue > prev_revenue
    GROUP BY customer_id
    HAVING COUNT(*) = (
        SELECT COUNT(*) - 1
        FROM monthly_revenue mr
        WHERE mr.customer_id = revenue_diff.customer_id
    )
)
SELECT rd.customer_id, rd.month, rd.revenue, rd.prev_revenue
FROM revenue_diff rd
WHERE rd.customer_id IN (SELECT customer_id FROM increasing_customers)
ORDER BY rd.customer_id, rd.month
"""
df = spark.sql(query)
display(df)
````

````python
%sql
-- Your solution:
WITH monthly_revenue AS (
    SELECT 
        customer_id,
        DATE_TRUNC('month', order_date) AS Order_Month, 
        SUM(amount) AS total_amount 
    FROM orders 
    GROUP BY customer_id, Order_Month
), 
revenue_diff AS (
    SELECT *, 
        LAG(total_amount) OVER (PARTITION BY customer_id ORDER BY Order_Month) AS prev_revenue 
    FROM monthly_revenue
)
SELECT 
    customer_id, 
    Order_Month,
    total_amount AS revenue, 
    prev_revenue 
FROM revenue_diff 
WHERE total_amount > prev_revenue
ORDER BY customer_id, Order_Month;
````
