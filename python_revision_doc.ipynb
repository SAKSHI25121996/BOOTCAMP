# 🧠 1️⃣ CORE DATA TYPES OVERVIEW

| Type  | Ordered | Mutable | Unique      | Use Case                |
| ----- | ------- | ------- | ----------- | ----------------------- |
| List  | ✅       | ✅       | ❌           | Store & iterate data    |
| Tuple | ✅       | ❌       | ❌           | Fixed combination (key) |
| Set   | ❌       | ✅       | ✅           | Remove duplicates       |
| Dict  | ❌       | ✅       | Keys unique | Mapping / grouping      |

# 🧩 2️⃣ LIST

## ✅ Create

```python
lst = [10, 20, 30]
```

## 🔍 Access

```python
lst[0]        # 10
for i in lst:
    print(i)
```

## ➕ Add

```python
lst.append(40)
lst.extend([50,60])
```

## ❌ Remove

```python
lst.remove(20)
lst.pop()      # last element
```

## 🎯 Use Cases

* Iteration
* Storing ordered data
* Transformations

# 🧩 3️⃣ DICTIONARY

## ✅ Create

```python
d = {"id": 1, "name": "A"}
```

## 🧠 Terminology

| Term  | Meaning  |
| ----- | -------- |
| Key   | "id"     |
| Value | 1        |
| Item  | ("id",1) |

---

## 🔍 Access

```python
d["id"]
for k, v in d.items():
    print(k, v)
```

## ➕ Add / Update

```python
d["age"] = 25
```

## ❌ Remove

```python
d.pop("id")
del d["name"]
```

## 🎯 Use Cases

* Grouping
* Counting
* Mapping

## 🧠 Breakdown

```text
data → list
item → dict
marks → list
```

# 🧠 7️⃣ COMMON PATTERNS

## 🔹 Grouping

```python
result = {}

for item in data:
    key = item["dept"]
    if key not in result:
        result[key] = []
    result[key].append(item["val"])
```
