````python
%sql
CREATE OR REPLACE TEMP VIEW orders AS
SELECT * FROM VALUES
 (101, '2024-01-05'),
 (101, '2024-02-10'),
 (101, '2024-04-01'),

 (102, '2024-01-15'),
 (102, '2024-02-20'),
 (102, '2024-03-10'),

 (103, '2024-02-05'),
 (103, '2024-03-08'),

 (104, '2024-01-25'),
 (104, '2024-03-15'),

 (105, '2024-03-12'),
 (105, '2024-04-18'),

 (106, '2024-04-10')
AS orders(customer_id, order_date);
````

````python
%sql

-- Find customers who ordered in **consecutive months**

-- 👉 Example:

-- * Jan + Feb → retained
-- * Feb + Mar → retained

with cte as (
select 
customer_id, date_trunc('Month', order_date) as order_month 
from orders
group by customer_id, date_trunc('Month', order_date)
),
cte2 as (
select 
customer_id, 
order_month,
lead(order_month, 1) over(partition by customer_id order by order_month) as next_month
from cte
)
select distinct
customer_id, 
order_month,
next_month,
MONTHS_BETWEEN(next_month, order_month) as diff
from cte2
where MONTHS_BETWEEN(next_month, order_month) = 1
````

````python
%sql

-- Calculate retention count per month
-- | month | retained_customers |

select  date_trunc('Month', order_date) as order_month , count(customer_id) from orders 
group by date_trunc('Month', order_date)
````

````python

# Count frequency of numbers
nums = [1,1,2,3,2,1,4]

# Output:
# {1:3, 2:2, 3:1, 4:1}

count_num = {}

for i in nums:
    if i in count_num:
        count_num[i] += 1
    else:
        count_num[i] = 1    
        
print(count_num)

````

````python
# Remove duplicates from list of dicts
# Input:

data = [
 {"id":1,"val":100},
 {"id":2,"val":200},
 {"id":1,"val":150}
]
# Output:

# [
#  {"id":1,"val":100},
#  {"id":2,"val":200}
# ]

seen = set()
result = []

for i in data:
    key = (i["id"], i["val"])
    if key not in seen:
        seen.add(key)
        result.append(i)
        
print(result)
# Flatten nested list
#])
````

````python
data = [
 {"id":1,"val":100},
 {"id":2,"val":200},
 {"id":1,"val":150}
]

# Method 1: sorted() - creates NEW list (doesn't modify original)
print("Method 1: sorted()")
top2_sorted = sorted(data, key=lambda x: x["val"], reverse=True)[:2]
print(top2_sorted)
print(f"Original data unchanged: {data}\n")

# Method 2: heapq.nlargest() - most efficient for small k
import heapq
print("Method 2: heapq.nlargest()")
top2_heap = heapq.nlargest(2, data, key=lambda x: x["val"])
print(top2_heap)
print()
````
