# 1️⃣ What is a Process?

A **process** is simply:

> A running instance of a program

### Example

* You run a Python script → that becomes a **process**
* You open Spark job → multiple **processes** run (driver + executors)

### Key points (interview-ready)

* Has its own:

  * Memory space
  * Resources (file handles, network, etc.)
* Is isolated from other processes

### Data Engineering angle

* In **Apache Spark**:

  * Driver = 1 process
  * Executors = multiple processes across nodes

👉 Good answer line:

> “A process is an isolated execution unit with its own memory space; in distributed systems like Spark, multiple processes work together across nodes.”

---

# 2️⃣ What is Memory?

Memory (RAM) is:

> Temporary storage where programs keep data they are actively using

### Think of it like:

* RAM = your desk (fast, limited)
* Disk = storage cabinet (slow, large)

### What lives in memory?

* Variables
* DataFrames
* Intermediate computations

### Data Engineering angle

* Spark tries to:

  * Keep data **in memory** for speed
  * Spill to disk if memory is insufficient

👉 Interview line:

> “Memory stores active data for fast access; in data engineering systems, efficient memory usage is critical to avoid spills and performance degradation.”

---

# 3️⃣ CPU vs RAM

This is one of the most important comparisons.

| Component | Role                                |
| --------- | ----------------------------------- |
| **CPU**   | Executes instructions (computation) |
| **RAM**   | Stores data being used              |

### Simple analogy

* CPU = chef 👨‍🍳
* RAM = kitchen counter 🍳

👉 More RAM → more ingredients ready
👉 Faster CPU → faster cooking

---

### Data Engineering angle

#### CPU-bound job

* Heavy transformations
* Complex joins
* UDFs

#### Memory-bound job

* Large datasets
* Caching
* Wide transformations

👉 Interview line:

> “CPU handles computation, while RAM holds working data; data pipelines can be CPU-bound or memory-bound depending on workload characteristics.”

---

# 1️⃣ Threads vs Processes

### 🔹 Process

* Independent execution unit
* Has its **own memory**
* Slower to create, but safer (isolated)

### 🔹 Thread

* Runs **inside a process**
* Shares the same memory with other threads
* Lightweight and faster

---

### Simple analogy

* Process = separate houses 🏠
* Threads = people inside the same house 👨‍👩‍👧‍👦

---

### Key difference

| Feature   | Process  | Thread |
| --------- | -------- | ------ |
| Memory    | Separate | Shared |
| Speed     | Slower   | Faster |
| Isolation | High     | Low    |

---

### Data Engineering angle

In **Apache Spark**:

* Each **executor = process**
* Inside executor → multiple **threads** run tasks

👉 Interview line:

> “Processes provide isolation, while threads enable parallelism within a process; Spark uses executors as processes and threads to run tasks concurrently.”

---

# 2️⃣ How Spark Uses Memory Internally

This is a *very strong* interview topic.

Spark divides memory into:

### 🔹 Execution Memory

Used for:

* Shuffles
* Joins
* Aggregations

### 🔹 Storage Memory

Used for:

* Caching
* Persisted DataFrames

---

### Important behavior

* If execution needs more memory → it can **take from storage**
* But storage cannot always take back from execution

---

### Real-world example

* You cache a large DataFrame
* Then run a join
* Spark may:

  * Evict cached data
  * Or spill to disk

---

👉 Interview line:

> “Spark dynamically shares memory between execution and storage; improper caching or large shuffles can lead to spills or OOM errors.”

---

# 3️⃣ Why Data Skew Causes Crashes

This is one of the **most asked real-world questions**.

### 🔹 What is Data Skew?

> Uneven distribution of data across partitions

---

### Example

* 1 partition → 90% of data
* Others → 10%

---

### What happens?

In **Apache Spark**:

* One executor gets overloaded
* That executor:

  * Runs longer ⏳
  * Uses more memory 💥
  * May crash (OOM)

---

### Symptoms

* One task takes much longer
* Executor failure
* Stage stuck at ~99%

---

### Fixes (interview gold)

#### ✅ Repartition

```python
df.repartition(100)
```

#### ✅ Salting (for joins)

* Add random key to distribute skew

#### ✅ Broadcast join

* If one table is small

#### ✅ Skew join optimization (Spark feature)

---

👉 Interview line:

> “Data skew leads to uneven workload distribution, causing some executors to run out of memory while others stay idle.”

# 1️⃣ Shuffle vs Narrow Transformations

### 🔹 Narrow Transformation

> Data stays within the same partition

Examples:

* `map()`
* `filter()`

👉 No data movement across nodes

---

### 🔹 Shuffle Transformation

> Data moves across partitions/nodes

Examples:

* `join()`
* `groupBy()`
* `reduceByKey()`

---

### Why shuffle is expensive

* Network I/O 🌐
* Disk spill 💾
* Serialization overhead

---

### Data Engineering angle (Spark)

In **Apache Spark**:

* Narrow → fast, pipeline execution
* Shuffle → creates new stage + heavy cost

---

👉 Interview line:

> “Narrow transformations are efficient as they avoid data movement, while shuffle transformations are expensive due to network and disk I/O.”

---

# 2️⃣ Broadcast Join vs Shuffle Join

This is asked *very frequently*.

---

## 🔹 Shuffle Join (default)

* Both tables are redistributed
* Happens when:

  * Tables are large

👉 Problem:

* Heavy shuffle
* Slow

---

## 🔹 Broadcast Join

> Small table is sent to all executors

```python
from pyspark.sql.functions import broadcast

df_large.join(broadcast(df_small), "id")
```

---

### Why it's fast

* No shuffle of large table
* Local joins happen in memory

---

### When to use

* One table is small (< ~10–100 MB typically)

---

👉 Interview line:

> “Broadcast join avoids shuffle by sending the smaller dataset to all executors, significantly improving join performance.”

---

# 3️⃣ Partitioning Strategies

This is where many candidates struggle—so let’s make it simple.

---

## 🔹 What is partitioning?

> Dividing data into chunks for parallel processing

---

## 🔹 Types

### ✅ Repartition

```python
df.repartition(100)
```

* Full shuffle
* Used to **increase parallelism**

---

### ✅ Coalesce

```python
df.coalesce(10)
```

* Reduces partitions
* Avoids full shuffle

---

### ✅ Partition by column (write time)

```python
df.write.partitionBy("date")
```

---

## 🔥 Why it matters

* Too few partitions → underutilized cluster
* Too many → overhead
* Bad partition key → data skew

---

👉 Interview line:

> “Proper partitioning ensures balanced workload and optimal resource utilization; poor partitioning leads to skew and performance issues.”
