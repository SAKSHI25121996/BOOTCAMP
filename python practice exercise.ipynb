# 🧠 PART 1 — Terminology (Simple & Clear)

You don’t need theory—just clarity.

## 🧩 Dictionary Terms

```python
d = {"id": 1, "name": "A"}
```

| Term                  | Meaning              |
| --------------------- | -------------------- |
| key                   | `"id"`, `"name"`     |
| value                 | `1`, `"A"`           |
| key-value pair (item) | `("id", 1)`          |
| dictionary            | whole `{...}` object |

👉 Important methods:

```python
d.keys()    # all keys
d.values()  # all values
d.items()   # (key, value) pairs
```

---

## 🧩 List Terms

```python
lst = [10, 20, 30]
```

| Term           | Meaning          |
| -------------- | ---------------- |
| element / item | 10, 20, 30       |
| index          | position (0,1,2) |

---

## 🧩 Nested Structure Example

```python
data = [
 {"name": "A", "marks": [10,20]}
]
```

Break it:

```text
data → list
data[0] → dictionary
data[0]["marks"] → list
```

👉 This is the MOST important understanding.

# 🧪 TEST 5 — Grouping

```python
data = [
 {"dept":"A","val":10},
 {"dept":"B","val":20},
 {"dept":"A","val":30}
]
```

### Task:

```python
{"A":[10,30], "B":[20]}
```

````python
data = [
 {"dept":"A","val":10},
 {"dept":"B","val":20},
 {"dept":"A","val":30}
]

test = {}

for i in data:
    dept = i['dept']
    if dept not in test:
        test[dept] = []
        test[dept].append(i['val']) 
    else:

        test[dept].append(i['val'])
print(test)
````

# 🧪 TEST 6 — Key-Based Dedup

```python
data = [
 {"id":1,"val":100},
 {"id":1,"val":200},
 {"id":2,"val":300}
]
```

### Task:

Keep only first occurrence of each `id`

---

````python
data = [
 {"id":1,"val":100},
 {"id":1,"val":200},
 {"id":2,"val":300}
]

result = []
set_dic = []
for i in data:
    id = i['id']
    if id not in result:
        set_dic.append({'id': id , 'val': i['val']})
        result.append(id)
    else:
        pass

print(set_dic)
````
